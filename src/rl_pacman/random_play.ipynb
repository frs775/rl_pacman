```python
from rl_pacman.environment import PacmanEnv
```

```python
# Create Environment
pacman_env = PacmanEnv()
pacman_env.display()

# Random PLay
num_steps = 0
while not pacman_env.done:
    num_steps += 1
    action = pacman_env.action_space.sample()
    pacman_env.move(action)
    if num_steps % 100 == 0:
        pacman_env.display()
        
# Display Final State
pacman_env.display()
```

```python
# Close ENvironment
pacman_env.close()
```
